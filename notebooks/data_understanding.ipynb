```python
import tensorflow as tf
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import seaborn as sns
```

```python
DATA_DIR = "../data/raw"

# Nur echte Unterordner als Klassen nehmen
classes = sorted([
    c for c in os.listdir(DATA_DIR)
    if os.path.isdir(os.path.join(DATA_DIR, c))
])
print("Klassen:", classes)
```

```python
# Nur Verzeichnisse als Klassen nutzen
classes = [c for c in classes if os.path.isdir(os.path.join(DATA_DIR, c))]

# Jetzt zählen wir pro Klasse
counts = {c: len(os.listdir(os.path.join(DATA_DIR, c))) for c in classes}
print(counts)

# Balkendiagramm
plt.figure(figsize=(6,4))
sns.barplot(x=list(counts.keys()), y=list(counts.values()))
plt.title("Anzahl Bilder je Klasse")
plt.ylabel("Anzahl")
plt.show()
```

```python
fig, axes = plt.subplots(len(classes), 3, figsize=(8, 4*len(classes)))
for i, cls in enumerate(classes):
    sample_imgs = os.listdir(os.path.join(DATA_DIR, cls))[:3]
    for j, fname in enumerate(sample_imgs):
        ax = axes[i, j]
        img = Image.open(os.path.join(DATA_DIR, cls, fname))
        ax.imshow(img)
        ax.axis("off")
        ax.set_title(cls)
plt.tight_layout()
plt.show()
```
